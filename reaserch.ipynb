```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data  
y = iris.target  
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python

tsne_configs = [
    {'init': 'pca', 'learning_rate': 200, 'perplexity': 50, 'early_exaggeration': 1.0, 'n_iter': 1000, 
     'n_iter_without_progress': 300, 'min_grad_norm': 1e-07, 'metric': 'dice', 'random_state': 42, 
     'verbose': 1, 'title': "t-SNE (learning_rate=200, perplexity=149, iter=1000)"},
    
    {'init': 'pca', 'learning_rate': 200, 'perplexity': 50, 'early_exaggeration': 12, 'n_iter': 1000, 
     'n_iter_without_progress': 300, 'min_grad_norm': 1e-07, 'metric': 'jaccard', 'random_state': 42, 
     'verbose': 1, 'title': "t-SNE (learning_rate=200, perplexity=0.000000001, iter=1000)"},
    
    {'init': 'pca', 'learning_rate': 200, 'perplexity': 50, 'early_exaggeration': 12.0, 'n_iter': 1000, 
     'n_iter_without_progress': 300, 'min_grad_norm': 1e-07, 'metric': 'canberra', 'random_state': 42, 
     'verbose': 1, 'title': "t-SNE (learning_rate=200, perplexity=0.1, iter=1000)"},
    
    {'init': 'pca', 'learning_rate': 200, 'perplexity': 50, 'early_exaggeration': 12.0, 'n_iter': 1000, 
     'n_iter_without_progress': 300, 'min_grad_norm': 1e-07, 'metric': 'sqeuclidean', 'random_state': 42, 
     'verbose': 1, 'title': "t-SNE (learning_rate=200, perplexity=50, iter=1000)"}
]

# Initialize the figure and axes for the subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Iterate through the configurations
for ax, config in zip(axes.flat, tsne_configs):
    tsne = TSNE(n_components=2, init=config['init'], learning_rate=config['learning_rate'], 
                perplexity=config['perplexity'], early_exaggeration=config['early_exaggeration'], 
                n_iter=config['n_iter'], n_iter_without_progress=config['n_iter_without_progress'],
                min_grad_norm=config['min_grad_norm'], metric=config['metric'], 
                random_state=config['random_state'])
    
    X_tsne = tsne.fit_transform(X)  
    
    sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y, palette='pastel', s=100, ax=ax)
    ax.set_title(config['title'])

plt.tight_layout()
plt.show()
```
